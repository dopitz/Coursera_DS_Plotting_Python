```python
import pandas as pd
import numpy as np
import numpy
import matplotlib.pyplot as plt
```

```python
#Reading phd file
doc = pd.read_csv('Matricula_Total_de_Doctorado_por_area_del_conocimiento.csv', sep=';', thousands = '.')
doc = doc.dropna(axis=0, how='all')
doc = doc.drop(doc.index[0])
doc.columns = ['Area', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011']
doc = doc.replace('-', 0)
doc = doc[:-1]


doc=doc.T
doc = doc[1:]
doc.columns = ['Agricultural and Marine Science', 'Art', 'Natural Science', 'Social Science', 'Law', 'Humanities', 'Education', 'Technology', 'Medical Science', 'TOTALES']
doc['YEAR'] = doc.index
doc.YEAR = doc.YEAR.astype(float)
doc.TOTALES = doc.TOTALES.astype(float)

doc=doc.apply(pd.to_numeric)

#For some values dots were considered as decimal. Fixing that

doc['TOTALES'] = doc['TOTALES'].apply(lambda x: x*1000 if x < 2 else x )

doc
```

```python
#Reading master file
mas = pd.read_csv('Matricula_Total_de_Magister_por_area_del_conocimiento.csv', sep=';', thousands = '.')
mas = mas.dropna(axis=0, how='all')
mas = mas.drop(mas.index[0])
mas.columns = ['Area', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011']
mas= mas[:-1]
mas = mas.T
mas = mas[1:]
mas.columns = ['Agricultural and Marine Science', 'Art', 'Natural Science', 'Social Science', 'Law', 'Humanities', 'Education', 'Technology', 'Medical Science', 'TOTALES']
mas.TOTALES = mas.TOTALES.astype(float)
mas['YEAR'] = mas.index
mas.YEAR = mas.YEAR.astype(float)
mas=mas.apply(pd.to_numeric)

mas
```

```python
DOC=doc.T

MAS=mas.T

DOC = DOC.iloc[:-2]

DOC

MAS = MAS.iloc[:-2]

MAS
```

```python
#Plotting lines

fig, ax = plt.subplots(2, 1, figsize=(10, 15)) 


doc.plot(x ='YEAR', y ='TOTALES', kind='line', ax=ax[0], color ='b', label ='Doctor', linewidth=4)
mas.plot(x ='YEAR', y ='TOTALES', kind='line', ax=ax[0], color ='r', label ='Master', linewidth=4)

ax[0].legend().set_visible(False)

ax[0].spines['top'].set_visible(False) 
ax[0].spines['right'].set_visible(False) 

ax[0].set_title('Students enrolled in 2001-2011 period', fontsize=16)
ax[0].set_xlabel('Year', fontsize=16)
ax[0].set_ylabel('NÂº Students', fontsize=16)


```

```python
#Creating nee data frame to plot bars
df = DOC[['2011']].copy()
df['Master'] = MAS['2011']
df = df.rename(columns= {'2011': 'Doctor'})
df
```

```python
#Plotting Bars
my_colors = ['b', 'r']
df.plot.bar(stacked=True, ax=ax[1], color=my_colors)

fig.suptitle('Graduate students enrolled in Chilean traditional universities', fontsize=18)

ax[0].tick_params(labelsize=16)
ax[1].tick_params(labelsize=16)
ax[1].set_title('Students enrolled in 2011 per area', fontsize=16)
ax[1].legend(fontsize=16)
ax[1].set_xlabel('Area', fontsize=16)
ax[1].set_ylabel('NÂº Students', fontsize=16)


ax[1].spines['top'].set_visible(False) 
ax[1].spines['right'].set_visible(False) 

plt.show()

fig.gca()
```
